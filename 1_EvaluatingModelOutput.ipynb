```python
train_items = glob.glob('./datasets/MIO-TCD-Classification/train/*/*')
val_items = glob.glob('./datasets/MIO-TCD-Classification/val/*/*')
test_items = glob.glob('./datasets/MIO-TCD-Classification/test/*/*')
```

```python
data_df = pd.DataFrame(train_items, columns = ['path'] )
data_df['path'] = data_df['path'].str.split('MIO-TCD-Classification/').str[1]
data_df['target'] = data_df['path'].str.split('/').str[1]
```

```python
def process_paths(items, dataset_name = 'MIO-TCD-Classification/'):
    data_df = pd.DataFrame(items, columns = ['path'] )
    # Split the 'path' column at 'MIO-TCD-Classification/' and keep only the second part
    data_df['path'] = data_df['path'].str.split(dataset_name).str[1]
    # Extract the target value as the second part of the split 'path' column based on '/'
    data_df['target'] = data_df['path'].str.split('/').str[1]
    return data_df
```

```python
train_df = process_paths(train_items)
val_df = process_paths(val_items)
test_df = process_paths(test_items)
```

```python
train_df.to_csv('train_data.csv')
val_df.to_csv('val_data.csv')
test_df.to_csv('test_data.csv')
```

```python
# Open the image file and encode it as a base64 string
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode("utf-8")

base64_image_1 = encode_image('class_0_1.png')
base64_image_2 = encode_image('class_3_1.png')
base64_image_3 = encode_image('class_3_2.png')
base64_image_4 = encode_image('class_0_2.png')

base64Frames = [base64_image_1, base64_image_2, base64_image_3, base64_image_4]
```

```python
import glob as glob
from random import sample 
sample(glob.glob(f'./datasets/MIO-TCD-Classification/val/5/*'), 3)
```

```python
def get_samples_for_class(class_folder, num_to_select, train_or_val = "train"):

    # Get the last two images in the folder
    samples_to_choose = sample(glob.glob(f'./datasets/MIO-TCD-Classification/{train_or_val}/{class_folder}/*'), num_to_select)

    base64Frames = []

    # Encode them
    for item in samples_to_choose:
        base64_image_1 = encode_image(item)
        base64Frames.append(base64_image_1)

    # Return as a list
    return base64Frames
```
